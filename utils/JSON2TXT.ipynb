# Notebook creado para convertir etiquetas de _polygon_ de labelme a formato txt

## Librerias

```python
import os
import json
import numpy as np
```

## Variables de ambiente

```python
PATH_ETIQUETAS = '/content/drive/MyDrive/segmentado/etiquetas/' # ruta donde estan los archivos .json de labelme
PATH_TXT = '/content/drive/MyDrive/segmentado/etiquetas_txt/' # ruta donde se guardar치n los archivos convertidos a txt
```

## Funci칩n para convertir formatos

```python
def json2txt(path_json,path_txt):
    with open(path_json,'r', encoding='gb18030') as path_json:
        jsonx=json.load(path_json)
        with open(path_txt,'w+') as ftxt:
            for shape in jsonx['shapes']:
                xy=np.array(shape['points'])
                label=str(shape['label'])
                strxy = ''
                for m,n in xy:
                    strxy+=str(m)+','+str(n)+','
                strxy+=label
                ftxt.writelines(strxy+"\n")
```

## Ejecuci칩n de algoritmo

```python
if not os.path.exists(PATH_TXT): # si no existe la carpeta se la debe crear
    os.makedirs(PATH_TXT)
list_json = os.listdir(PATH_ETIQUETAS)
for cnt,json_name in enumerate(list_json):
    print('cnt=%d,name=%s'%(cnt,json_name))
    path_json = PATH_ETIQUETAS + json_name
    path_txt = PATH_TXT + json_name.replace('.json','.txt')
    # print(path_json, path_txt)
    json2txt(path_json, path_txt) # llamando a funci칩n que convierte formatos
```
